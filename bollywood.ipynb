```python
final['comb'] = final['Title'] + " " + final['Director'] + " " + final['Cast'] + " " + final['Genre']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
comb = cv.fit_transform(final['comb']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(comb)
```

```python
def get_title_from_index(index):
    return final[final.index == index]['Title'].values[0]

def get_index_from_title(title):
    return final[final.Title==title].index[0]
```

```python
def return_similar_movies(movie):
    movie = movie.capitalize()
    movie_index = get_index_from_title(movie)
    
    similar_movies = list(enumerate(cosine_sim[movie_index]))
    
    sorted_similar_movies = sorted(similar_movies ,key=lambda x:x[1],reverse=True)
    
    output = []
    for movie in sorted_similar_movies:
        output.append(get_title_from_index(movie[0]))
        
    return output[1:20]
        
    
    
    
```
